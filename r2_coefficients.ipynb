```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,1.1,1.25,1.5,2,2.2,3.5,5,6,6.5,7,8,8.75,9.5,10])
y = np.array([10,9,8.9,7,6,5.5,4,2,1.5,1.1,1.05,1,1,1.1,0.95])

#retorna o r2 de um polinômio
def r2 (polinomio, grau: int):
    g_de_x = np.array([0])
    for i in range(grau+1):
        aux = (polinomio[i] * pow(x,(grau-i)) )
        aux2 = g_de_x + aux
        g_de_x = aux2
    residuo = g_de_x - y
    SQRes = sum( pow(residuo,2) )
    total = y -(sum(y)/len(y))
    SQTot = sum( pow(total,2) )
    resp = 1 - (SQRes/SQTot)
    return  resp

polinomios = []
n_pol = len(x)

#armazena os polinômios calculados
for i in range(n_pol):
    polinomios.append(np.polyfit(x,y,i))

    #Coeficiente de Determinação R²:
    
#armazena os valores de r2 de cada polinômio e imprime-os na tela
vetor_r2 = []
print("Valores do Coeficiente de determinação '('R²')':\n")
for i in range(n_pol):
    vetor_r2.append( r2(polinomios[i],i) )
    if i == 0:
        pass
    elif i == 1:
        print( "Reta: "+ str(vetor_r2[i]) )
    else:
        print( "Polinômio de grau "+ str(i) + ": " + str(vetor_r2[i]) )

#plota o gráfico com os pontos e o polinômio de maior grau
plt.plot(x,y,'o')
plt.plot(x,np.polyval(polinomios[n_pol-1],x),'b-')

#OBS: polinomios de grau maior que 14 causam warning da função que calcula os polionomios
```
