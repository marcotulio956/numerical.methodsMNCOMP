```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.5,1,2,3,4])
y = np.array([2500, 2400, 2000, 1800, 1500])

#A produção (y) cai linearmente,
#o que sugere uma equação linear para o comportamento dos dados
reta = np.polyfit(x,y,1)

#plota o gráfico com os pontos e a reta calculada
plt.plot(x,y,'o')
plt.plot( x,np.polyval(reta,x ),'b-')


#calculo do valor de x para y = 500:

    #g(x) = a * x_resposta + b,
    #no qual a = reta[0] e b = reta[1]
    #rearranjando:
    
x_resposta = (500 - reta[1])/reta[0]

x_resposta
```
